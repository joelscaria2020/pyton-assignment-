```
stud = pd.read_csv ("Student Performance.csv")
stud.head()
```

```
mthscr = np.array(stud.iloc[:,[5]].values)
rdscr = np.array(stud.iloc[:,[6]].values)
```

```
print("mthscr",mthscr[:5])
print("rdscr") ,rdscr[:6]
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(mthscr ,rdscr)
model.score(mthscr,rdscr)
```

```
import matplotlib.pyplot as plt
plt.scatter(mthscr , rdscr)
print(mthscr.shape)
plt.plot(mthscr, model.predict(mthscr.reshape(-1,1)), 'r')
```
